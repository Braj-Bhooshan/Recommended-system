```python
# import library
import numpy as np
import pandas as pd
```

```python
# load dataset
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# view top 10 rows
movies.head()
```

```python
# print dataset shape
print(movies.shape)
print(credits.shape)
```

```python
# merge movies dataset and credits dataset in one file "movies"
movies = movies.merge(credits,on='title')
```

```python
# Shape after merge dataset

movies.shape
```

```python
# Removing colunms based on our understanding (which column put on my dataset)
#1. genres
#2. id
#3. keywords
#4. title
#5. overview
#6. cast
#7 crew
```

```python
# Filtering required column and stor all data in movies dataset
movies = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
# make string list dictionary to list dictionary
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast']= movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
movies['crew'][4808]
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```
